## Thresholding experiments
Compute the best accuracy for different thresholds.

Using only 1 iteration:

```python
segmentation_it1 = 'SEG_it1'
reference = 'gt_15_classes.nii.gz'
th_list = np.arange(0.1, 0.95, 0.05)
stats_th_it1 = compute_stats_by_threshold(options, segmentation_it1, reference, th_list)
```

```python
stats_th_it1 = np.array(stats_th_it1)
dsc_all_th = np.array([[item.values()[0][0] for item in t] for t in stats_th_it1])
hd_all_th = np.array([[item.values()[0][1] for item in t] for t in stats_th_it1])

print dsc_all_th.shape
means = np.mean(np.mean(dsc_all_th, axis = 1), axis = 1)
print dsc_all_th.shape
means = np.mean(np.mean(dsc_all_th, axis = 1), axis = 1)
print means
print 'best_th: ', th_list[np.argmax(means)], ':', means[np.argmax(means)]
```

Using a cascade of two classifiers

```python
segmentation_cascade = 'SEG_cascade'
reference = 'gt_15_classes.nii.gz'
th_list = np.arange(0.1, 0.95, 0.05)
stats_th_cascade = compute_stats_by_threshold(options, segmentation_cascade, reference, th_list)
```

```python
stats_th_cascade = np.array(stats_th_cascade)
dsc_all_th = np.array([[item.values()[0][0] for item in t] for t in stats_th_cascade])
hd_all_th = np.array([[item.values()[0][1] for item in t] for t in stats_th_cascade])

print dsc_all_th.shape
means = np.mean(np.mean(dsc_all_th, axis = 1), axis = 1)
print means
print 'best_th: ', th_list[np.argmax(means)], ':', means[np.argmax(means)]
```
